## Step 3: Data Preparation

```python
## dropping columns that are irrelevant for the model - 
df.drop(['instant','dteday','casual','registered'],axis = 1,inplace = True)
df.head()
```

```python
### Converting some numeric values to categorical data
import calendar
df['mnth'] = df['mnth'].apply(lambda x: calendar.month_abbr[x])
```

```python
## maping seasons
df.season = df.season.map({1: 'Spring',2:'Summer',3:'Fall',4:'Winter'})
```

```python
## mapping weathersit
df.weathersit = df.weathersit.map({1:'Clear',2:'Mist & Cloudy', 
                                             3:'Light Snow & Rain',4:'Heavy Snow & Rain'})
```

```python
## mapping weekday
df.weekday = df.weekday.map({0:"Sunday",1:"Monday",2:"Tuesday",3:"Wednesday",4:"Thrusday",5:"Friday",6:"Saturday"})
```

```python
df.head()
```

### Dummy Variables

```python
### creating dummy variables for season , mnth ,weathersit ,weekday
dummy = df[['season','mnth','weekday','weathersit']]
```

```python
dummy = pd.get_dummies(dummy,drop_first=True )
```

```python
## adding dummy variables to original dataset
df = pd.concat([dummy,df],axis = 1)
df.head()
```

```python
## dropping columns for which dummy variables were created
df.drop(['season', 'mnth', 'weekday','weathersit'], axis = 1, inplace = True)
df.head()
```

```python
df.shape
```

## Step 4: Splitting the Data into Training and Testing Sets

As you know, the first basic step for regression is performing a train-test split.

```python
train, test = train_test_split(df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

## Step 5 :Rescaling the Features 

```python
scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the columns except the 'dummy' variables.
num_vars = ['cnt','hum','windspeed','temp','atemp']

train[num_vars] = scaler.fit_transform(train[num_vars])
train.head()
```

```python
train.describe()
```

```python
plt.figure(figsize = (20, 10))
sns.heatmap(train.corr(), annot = True, cmap="YlGn")
plt.show()
```

### Dividing into X and Y sets for the model building

```python
y_train = train.pop('cnt')
X_train = train
```

## Step 6: Building a linear model

Fit a regression line through the training data using `statsmodels`. In `statsmodels`, you need to explicitly fit a constant using `sm.add_constant(X)` because if we don't perform this step, `statsmodels` fits a regression line passing through the origin, by default.

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 13)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
#Columns selected by RFE and their weights
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

### Starting with all the columns selected by RFE

### Model 1

```python
#Print Columns selected by RFE. We will start with these columns for manual elimination
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
 
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
# Create a first fitted model
lm = sm.OLS(y_train,X_train_rfe).fit()
```

```python
# Check the parameters obtained

lm.params
```

```python
# Print a summary of the linear regression model obtained
print(lm.summary())
```

dropping weekday_Saturday since it has p > 0.05

```python
X_train_new = X_train_rfe.drop(["weekday_Saturday"], axis = 1)
```

Rebuilding the model without 'weekday_Saturday'

### Model 2

```python
# Adding a constant variable 
X_train_lm = sm.add_constant(X_train_new)
```

```python
lm = sm.OLS(y_train,X_train_lm).fit()   # Running the linear model
```

```python
print(lm.summary())
```

### Checking VIF for multicollinearity

```python
# Calculate the VIFs for the new model
vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

#### We generally want a VIF that is less than 5. So there are clearly some variables we need to drop.


```python
X_train_new = X_train_new.drop(['const'], axis=1)
```

```python
# Calculate the VIFs for the new model again
vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### Model 3 (final model)

```python
### dropping hum from the model
X_train_new = X_train_new.drop(['hum'], axis=1)
```

```python
# Adding a constant variable 
X_train_lm = sm.add_constant(X_train_new)

# Create a first fitted model
lm = sm.OLS(y_train,X_train_lm).fit()   
```

```python
# Check the summary
print(lm.summary())
```

```python
# Calculate the VIFs for the new model again
vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

## Step 7: Residual Analysis of the train data

So, now to check if the error terms are also normally distributed (which is infact, one of the major assumptions of linear regression), let us plot the histogram of the error terms and see what it looks like.

```python
y_train_cnt = lm.predict(X_train_lm)
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_cnt), bins = 15)
fig.suptitle('Error Terms', fontsize = 15)                   
plt.xlabel('Errors', fontsize = 16)                         
```

## Step 8: Making Predictions Using the Final Model

Now that we have fitted the model and checked the normality of error terms, it's time to go ahead and make predictions using the final, i.e. third model.

#### Applying the scaling on the test sets

```python
num_vars = ['cnt','hum','windspeed','temp','atemp']
test[num_vars] = scaler.transform(test[num_vars])
test.describe()
```

#### Dividing into X_test and y_test

```python
y_test = test.pop('cnt')
X_test = test
```

```python
# Adding constant variable to test dataframe
X_test = sm.add_constant(X_test)
```

```python
# predicting using values used by the final model
test_col = X_train_lm.columns
X_test=X_test[test_col[1:]]
# Adding constant variable to test dataframe
X_test = sm.add_constant(X_test)

X_test.info()
```

```python
# Making predictions using the fourth model

y_pred = lm.predict(X_test)
```

## R^2 Value for TEST


```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

## Adjusted R^2 Value for TEST

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
mse
```
